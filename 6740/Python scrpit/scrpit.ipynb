```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# กำหนดชื่อไฟล์
file_path = 'Pulse.csv'
```

```python
# อ่านไฟล์ CSV
df = pd.read_csv(file_path)
df
```

```python
# --- การเตรียมข้อมูล ---
# ทำความสะอาดชื่อคอลัมน์ โดยลบช่องว่างที่ไม่จำเป็นออก
df.columns = df.columns.str.strip()
```

```python
# --- การสร้างกราฟ ---
# กำหนดขนาดของกราฟ
plt.figure(figsize=(15, 7))

# พล็อตกราฟของ ENCODER A และ B เทียบกับเวลา
# แปลงค่า TRUE/FALSE เป็น 1/0 เพื่อให้พล็อตกราฟได้
# ใช้ drawstyle='steps-post' เพื่อให้เห็นการเปลี่ยนแปลงของสัญญาณดิจิทัลชัดเจน
plt.plot(df['time'], df['ENCODER  A:1'].astype(int), drawstyle='steps-post', label='Encoder A')
plt.plot(df['time'], df['ENCODER  B:1'].astype(int), drawstyle='steps-post', label='Encoder B')

# --- การตกแต่งกราฟ ---
# เพิ่มชื่อกราฟและชื่อแกน
plt.title('กราฟแสดงสัญญาณ Encoder เทียบกับเวลา')
plt.xlabel('เวลา (วินาที)')
plt.ylabel('สถานะสัญญาณ')

# กำหนดให้แกน Y แสดงเป็น False และ True
plt.yticks([0, 1], ['False', 'True'])

# เพิ่มเส้นตารางและคำอธิบายกราฟ (Legend)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# ปรับ layout ให้พอดี
plt.tight_layout()

# บันทึกรูปกราฟเป็นไฟล์ PNG
plt.savefig('Pulse_Signal_Plot.png')

print("กราฟถูกบันทึกเป็นไฟล์ 'Pulse_Signal_Plot.png' เรียบร้อยแล้ว")

# แสดงกราฟ
plt.show()
```
